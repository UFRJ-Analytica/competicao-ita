```python
from geopy.distance import geodesic

# Coordenadas dos aeroportos (latitude, longitude)
coordenadas = {
    'SBBR': (-15.8697, -47.9207),  # Brasília
    'SBFL': (-27.6703, -48.5525),  # Florianópolis
    'SBCT': (-25.5282, -49.1758),  # Curitiba
    'SBRJ': (-22.9102, -43.1631),  # Rio de Janeiro (Santos Dumont)
    'SBCF': (-19.6328, -43.9634),  # Belo Horizonte (Confins)
    'SBPA': (-29.9936, -51.1770),  # Porto Alegre
    'SBSP': (-23.6261, -46.6566),  # São Paulo (Congonhas)
    'SBRF': (-8.1264, -34.9240),   # Recife
    'SBSV': (-12.9105, -38.3311),  # Salvador
    'SBKP': (-23.0074, -47.1345),  # Campinas (Viracopos)
    'SBGR': (-23.4322, -46.4695),  # São Paulo (Guarulhos)
    'SBGL': (-22.8076, -43.2506)   # Rio de Janeiro (Galeão)
}

# Calcular distância entre todos os pares de aeroportos
for aeroporto1 in coordenadas:
    for aeroporto2 in coordenadas:
        if aeroporto1 != aeroporto2:
            distancia = geodesic(coordenadas[aeroporto1], coordenadas[aeroporto2]).kilometers
```

```python
def distancia(row):
     return geodesic(coordenadas[row['origem']], coordenadas[row['destino']]).kilometers

df['distancia'] = df.apply(distancia, axis=1)
df
```

```python
from metar import Metar
```

- temp

- wind_speed

- press
- wind_speed_peak
- press_sea_level
- precip_1hr 

```python
def download(row, met, tipo):
    meta = row[tipo]
    if tipo == 'metaf':
        meta = meta.replace('METAF', 'METAR').split('\n')[0]

    obs = Metar.Metar(meta)
    metodo = getattr(obs, met, None)
    if metodo == None:
        print(f'Id: {row.flightid}\nParametro: {met}')
        return 0
    return metodo.value()

def extracao_meta(df, tipo = 'metar'):
    sufixo = '_'+ tipo
    dados = [
    "temp",
    "vis",
    "wind_speed",
    "dewpt",
    "press",
    ]
    for met in dados:
        df[met+sufixo] = df.apply(download, axis=1, args=(met,tipo,))

    return df.copy()
```

```python
train = extracao_meta(df, 'metar')
```
